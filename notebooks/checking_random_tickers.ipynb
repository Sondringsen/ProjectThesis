<h1>Checking random tickers</h1>
<h3>This is done to ensure that the metadata of the random tickers are representative of the entire dataset</h3>

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
meta_data = pd.read_csv("../data/raw/meta_data.csv", index_col=0)
meta_data = meta_data.set_index(meta_data["ticker"])

exclude_columns = ["ticker"]
categorical_columns = ["sectorKey", "industryKey"]
# numerical_columns = meta_data.columns.drop(categorical_columns)
numerical_columns = meta_data.columns.drop(categorical_columns + exclude_columns)

# Removing infinite values
meta_data[numerical_columns] = meta_data[numerical_columns].replace([np.inf, -np.inf], np.nan)

random_tickers = pd.read_csv("../data/processed/tq_data_gridded/df_tot_gridded.csv", index_col=0)
random_tickers = random_tickers.index.unique()[:50]

meta_data_random = meta_data.loc[random_tickers, :]
```

<h3>T-statistics for all columns</h3>

```python
t_stat = (meta_data[numerical_columns].mean() - meta_data_random[numerical_columns].mean()) / meta_data[numerical_columns].std()
t_stat
```

<h3>Plotting histograms of all data and random subset data</h3>

```python
fig, axs = plt.subplots(numerical_columns.shape[0], figsize=(10, 50))

fig.suptitle("Histograms of columns")

for i, col in enumerate(numerical_columns):
    meta_data_col = meta_data[col].dropna()
    meta_data_col = meta_data_col.map
    axs[i].hist(meta_data[col].dropna(), alpha=0.5, label=f'{col} (full)', density=True, bins=20)
    axs[i].hist(meta_data_random[col], alpha=0.5, label=f'{col}, (random)', density=True, bins=20)
    axs[i].legend()

fig.tight_layout()
```
